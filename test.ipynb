```python
import requests
import pandas as pd
import datetime
import time
import random
```

```python
# 建立時間
begin = datetime.datetime(2010, 1, 1)
end = datetime.datetime.now() 
step = datetime.timedelta(days=365)
DatePeriod = []

while begin <= end:
    DatePeriod.append(begin.strftime('%Y%m%d'))
    begin += step
    
#計算到今年
DatePeriod.append(datetime.datetime.now().strftime('%Y%m%d'))
```

```python
dataframe = pd.DataFrame()
for year in DatePeriod:
    url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date='+ year +'&stockNo=2330'
    r = requests.get(url)
    print(r.status_code)
    dataframe = dataframe.append(pd.DataFrame(r.json()['data'], columns=r.json()['fields']))
    # df = pd.DataFrame(columns=["DATE", "TV", "TO", "OPEN", "HIGH", "LOW", "CLOSE", "PFL", "NT"])
    
    # 避免被黑名單
    time.sleep(random.randint(20, 60))
```
